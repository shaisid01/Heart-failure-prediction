```python
#Count the number of missing values in each column and sort them.
missing = pd.concat([df_hf.isnull().mean(), 100 * df_hf.isnull().mean()],axis=1)
missing.columns=['count','%']
missing.sort_values(by= 'count', ascending=False)
```
